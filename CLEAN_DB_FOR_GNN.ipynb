```python
import pandas as pd
df_train = pd.read_csv('./train.txt', header=None, names=['h','r','t'])
df_test = pd.read_csv('./test.txt', header=None, names=['h','r','t'])
```

```python
set(df_train.h.unique()).difference(df_test.h.unique())
```

```python
import numpy as np
train = df_train.h.unique().tolist() + df_train.t.unique().tolist()
test = df_test.h.unique().tolist() + df_test.t.unique().tolist()
diff_1 = set(train).difference(test)
print(len(diff_1), 100*len(diff_1)/float(len(set(train))))
```

```python
diff_2 = set(test).difference(train)
print(len(diff_2), 100*len(diff_2)/float(len(set(test))))
```
