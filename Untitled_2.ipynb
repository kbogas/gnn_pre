```python
from IPython.display import display
from ipywidgets import widgets
import numpy as np
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve, confusion_matrix
from sklearn.metrics import average_precision_score, auc, precision_recall_curve, roc_curve
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
import ipywidgets as ipw
from ipywidgets import *
from sklearn.externals.funcsigs import signature


font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)

SMALL_SIZE = 14
MEDIUM_SIZE = 22
BIGGER_SIZE = 22

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title


tr_slider = widgets.FloatSlider(
    value=0.5,
    min=0,
    max=1.01,
    step=0.01,
    description='Threshold:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.2f',
)



def tresh(x, tr=0.5):
    res = []
    for x_ in x:
        if x_ > tr:
            res.append(1)
        else:
            res.append(0)
    return np.array(res)

def plot_prec_rec_at_k(y_true, y_probas, k_max=1000):
    temp = y_probas.argsort()[::-1]
    cur_rank = np.empty_like(temp)
    cur_rank[temp] = np.arange(len(y_probas))
    logs = []
    tresh = 0.5
    for i, prob in enumerate(y_probas):
        #print(prob, y_true[i])
        if prob > tresh and y_true[i] == 1:
            kind = "True Positive"
        elif prob > tresh and y_true[i] == 0:
            kind = "False Positive"
        elif prob <= tresh and y_true[i] == 0:
            kind = 'True Negative'
        else:
            kind = 'False Negative'
        logs.append({"Rank":cur_rank[i], "Proba":prob, "Kind":kind})
    df = pd.DataFrame(logs)
    df = df.sort_values('Rank')
    df.reset_index(inplace=True)
    if len(df) <= k_max:
        k_max = len(df)
    prec = []
    rec = []
    ks = np.arange(1,k_max)
    for k_i in ks:
        cur_df = df.iloc[:k_i]
        tp = cur_df[cur_df['Kind']=="True Positive"].shape[0]
        fp = cur_df[cur_df['Kind']=="False Positive"].shape[0]
        fn = cur_df[cur_df['Kind']=="False Negative"].shape[0]
        prec.append(tp/float(tp+fp+0.001))
        rec.append(tp/float(tp+fn+0.001))
    plt.figure(figsize=(20,10))
    plt.rc('figure', titlesize=SMALL_SIZE)
    plt.plot(ks, prec, 'b--', label='precision')
    plt.plot(ks, rec, 'g--', label = 'recall')
    plt.legend(loc='upper right')
    plt.ylim([-0.01,1.05])
    plt.xlim([1,k_max])
    plt.rc('figure', titlesize=SMALL_SIZE)
    plt.title("Prec-Rec at up to %d " % (k_max))
    return

def plot_bars(y_true, y_probas, N=100):
    import pandas as pd
    import seaborn as sns
    plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
    plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
    plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
    plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
    plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
    temp = y_probas.argsort()[::-1]
    cur_rank = np.empty_like(temp)
    cur_rank[temp] = np.arange(len(y_probas))
    logs = []
    tresh = 0.5
    for i, prob in enumerate(y_probas):
        #print(prob, y_true[i])
        if prob > tresh and y_true[i] == 1:
            kind = "True Positive"
        elif prob > tresh and y_true[i] == 0:
            kind = "False Positive"
        elif prob <= tresh and y_true[i] == 0:
            kind = 'True Negative'
        else:
            kind = 'False Negative'
        logs.append({"Rank":cur_rank[i], "Proba":prob, "Kind":kind})
    df = pd.DataFrame(logs)
    df = df.sort_values('Rank')
    df.reset_index(inplace=True)
    palette ={"True Positive":"g",
              "True Negative":"b",
              "False Positive":"r", 
              "False Negative":"y"}
    legend_names = []
    legend_handles = []
    fig = plt.figure(figsize=(30,20))
    for i, row in df.iloc[:N].iterrows():
        h = plt.bar(i+1, row['Proba'], color=palette[row['Kind']], width=1, align='center')
        if not(row['Kind'] in legend_names):
            legend_names.append(row['Kind'])
            legend_handles.append(h)
    #plt.xticks([i+1 for i in range(len(df))], [str(i+1) for i in range(len(df))], rotation=45)
    plt.xlabel("Ranks")
    #plt.tick_params(axis='x', which='both',)
    plt.grid(axis='x')
    plt.ylim([0,1])
    plt.ylabel("Probability")
    ax = plt.axhline(tresh, color='w', linestyle='-.', lw=1)
    legend_names.append("Threshold %0.2f" % tresh)
    #legend_handles.append(ax)
    plt.legend(legend_handles, legend_names)
    plt.title("Probability of top-%d predictions vs their rank" % N)
    plt.show()

def plot_prec_recall_vs_tresh(precisions, recalls, thresholds, pos_label):
#     if pos_label == 0:
#         plt.subplot(1, 2, 1)
#         plt.plot(thresholds, precisions[:-1], 'b--', label='precision')
#         plt.plot(thresholds, recalls[:-1], 'g--', label = 'recall')
#         plt.plot(thresholds, 2*precisions[:-1]*recalls[:-1]/(precisions[:-1]+recalls[:-1]), 
#                  'r--', label = 'F-1')
#     else:
    plt.subplot(1, 2, 1)
    
    plt.plot(thresholds, precisions[:-1], 'b--', label='precision')
    plt.plot(thresholds, recalls[:-1], 'g--', label = 'recall')
    plt.plot(thresholds, 2*precisions[:-1]*recalls[:-1]/(precisions[:-1]+recalls[:-1]), 
             'r--', label = 'F-1')
    plt.xlabel('Threshold')
    plt.legend(loc='upper right')
    plt.ylim([-0.01,1.05])
    plt.xlim([-0.01,1.05])
    plt.axvline(0.5, color='y', linestyle='-.')
    plt.rc('figure', titlesize=SMALL_SIZE)
    plt.title("Prec/Rec/F1 for class %s vs treshold" % (str(pos_label)))

    
def Find_Optimal_Cutoff(fpr,tpr,threshold):
    i = np.arange(len(tpr)) 
    roc = pd.DataFrame({'tf' : pd.Series(tpr-(1-fpr), index=i), 'threshold' : pd.Series(threshold, index=i)})
    roc_t = roc.ix[(roc.tf-0).abs().argsort()[:1]]
    return list(roc_t['threshold'])[0]





def plot_roc_auc(y_test, preds, 
                 annotate_optimal=False, 
                 annotate_half=False,
                save_fig=False,
                save_name='ROC_AUC.pdf'):
    fig = plt.figure(figsize=(20,10))
    fpr, tpr, thresholds = roc_curve(y_test, preds)
    roc_auc = auc(fpr, tpr)
    plt.subplot(1,2,1)
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    if annotate_optimal:
        optimal = Find_Optimal_Cutoff(fpr, tpr, thresholds)
        optimal_index = np.where(thresholds>=optimal)[0][-1]
        plt.annotate('%0.2f' % optimal, (fpr[optimal_index], tpr[optimal_index]),     
            xycoords='data',
            xytext=(fpr[optimal_index], tpr[optimal_index]-0.1),
            textcoords='data',
            horizontalalignment='center',
            arrowprops=dict(facecolor='black', arrowstyle="->"))
    if annotate_half:
        half_index = np.where(thresholds>=0.5)[0][-1]
        plt.annotate('%0.2f' % 0.5, (fpr[half_index], tpr[half_index]),     
            xycoords='data',
            xytext=(fpr[half_index], tpr[half_index]-0.1),
            textcoords='data',
            horizontalalignment='center',
            arrowprops=dict(facecolor='black', arrowstyle="->"))
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
#     plt.title('Receiver Operating Characteristic')
#     plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
#     plt.legend(loc = 'lower right')
#     plt.plot([0, 1], [0, 1],'r--')
#     plt.xlim([0, 1])
#     plt.ylim([0, 1])
#     plt.ylabel('True Positive Rate')
#     plt.xlabel('False Positive Rate')
    plt.subplot(1,2,2)
    plt.title('Precision-Recall Curve')
    from sklearn.metrics import average_precision_score
    average_precision = average_precision_score(y_test, preds)

    precision, recall, _ = precision_recall_curve(y_test, preds)

    # In matplotlib < 1.5, plt.fill_between does not have a 'step' argument
    step_kwargs = ({'step': 'post'}
                   if 'step' in signature(plt.fill_between).parameters
                   else {})
    plt.step(recall, precision, color='b', alpha=0.2,
             where='post')
    plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([-0.05, 1.05])
    plt.xlim([-0.05, 1.05])
    plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(
              average_precision))
    if save_fig:
        plt.savefig('./Results/ROC_AUCS/%s.pdf' % save_name, dpi=320,bbox_inches='tight', pad_inches=0)
    
#    plot_roc(fpr, tpr, thresholds, label_every=30)

    
def plot_probas(all_preds_view):
    ax = plt.subplot(1,2,2)
    plt.rc('figure', titlesize=SMALL_SIZE)
    sns.distplot(all_preds_view.Probability.values, kde=True, bins=50, norm_hist=True)
    #grid = sns.FacetGrid(all_preds_view, hue="True", height=7, )
    #grid.map(sns.kdeplot, 'Probability', ax=ax)
    #grid.add_legend()
    plt.axvline(0.5, 0,5)
    plt.title("Distribution of probabilities for Predicted Interactions")
    plt.xlabel("Normalized frequency")
    plt.xlabel("Probability")

def interact_(tr_slider):
    from collections import Counter
   
    y_true = all_preds_view.True.values
    y_new = all_preds_view.Probability.values
    plt.figure(figsize=(20,8))
    true_c = Counter(y_true)[1]
    false_c = Counter(y_true)[0]
    plt.suptitle("Result for selected subset (True/False/Total): %d/%d/%d" % (true_c, false_c, true_c + false_c) )
#     pos_label = 0
#     prec, rec, tre = precision_recall_curve(y_true, y_new, pos_label=pos_label)
#     print(prec)
#    fig1 = plot_prec_recall_vs_tresh(prec, rec, tre, pos_label)
    pos_label = 1
    prec, rec, tre = precision_recall_curve(y_true, y_new, pos_label=pos_label)
#    print(prec)
    fig2 = plot_prec_recall_vs_tresh(prec, rec, tre, pos_label)
    plot_probas(all_preds_view)
    topN = 500
    plot_bars(y_true, y_new, N=topN)
    plot_roc_auc(y_true, y_new, False, False, save_fig=False)
    plot_prec_rec_at_k(y_true, y_new, k_max=topN)
    plt.show()
    
    cur_tresh = tr_slider
    y_class = tresh(y_new, tr=cur_tresh)
    print("Report")
    print classification_report(y_true, y_class)
    print("Confusion Matrix (rows are truth, cols are prediction)")
    print confusion_matrix(y_true, y_class)
    return 
all_preds_view = data
w_box = interactive(interact_,  tr_slider=tr_slider)
w_box
```

```python
import numpy as np
def negsamp_vectorized_bsearch_preverif(pos_inds, n_items, n_samp=32):
    """ Pre-verified with binary search
    `pos_inds` is assumed to be ordered
    """
    raw_samp = np.random.randint(0, n_items - len(pos_inds), size=n_samp)
    pos_inds_adj = pos_inds - np.arange(len(pos_inds))
    neg_inds = raw_samp + np.searchsorted(pos_inds_adj, raw_samp, side='right')
    return neg_inds
```

```python
import numpy as np

"""
This file implements various first-order update rules that are commonly used for
training neural networks. Each update rule accepts current weights and the
gradient of the loss with respect to those weights and produces the next set of
weights. Each update rule has the same interface:
def update(w, dw, config=None):
Inputs:
  - w: A numpy array giving the current weights.
  - dw: A numpy array of the same shape as w giving the gradient of the
    loss with respect to w.
  - config: A dictionary containing hyperparameter values such as learning rate,
    momentum, etc. If the update rule requires caching values over many
    iterations, then config will also hold these cached values.
Returns:
  - next_w: The next point after the update.
  - config: The config dictionary to be passed to the next iteration of the
    update rule.
NOTE: For most update rules, the default learning rate will probably not perform
well; however the default values of the other hyperparameters should work well
for a variety of different problems.
For efficiency, update rules may perform in-place updates, mutating w and
setting next_w equal to w.
"""


def sgd(w, dw, config=None):
    """
    Performs vanilla stochastic gradient descent.
    config format:
    - learning_rate: Scalar learning rate.
    """
    if config is None: 
        config = {}
        config.setdefault('learning_rate', learning_rate)
    print(w.shape, dw.shape)
    w = w - config['learning_rate'] * dw
    return w, config

def sgd_noise(w, dw, config=None):
    """
    Performs noisy sgd according to
    https://towardsdatascience.com/only-numpy-implementing-and-comparing-gradient-descent-optimization-algorithms-google-brains-8870b133102b
    config format:
    - learning_rate: Scalar learning rate.
    """
    if config is None: 
        config = {}
        config.setdefault('learning_rate', 0.001)
        config.setdefault('t', 0)
        
    config['t'] = config['t'] + 1
    add_noise_std = 1/np.power((1 + config['t']), 0.55)
    add_noise_value = np.random.normal(loc=0, scale=add_noise_std)
    w = w - config['learning_rate'] * (dw + add_noise_value)
    return w, config

def sgd_momentum(w, dw, config=None):
    """
    Performs stochastic gradient descent with momentum.
    config format:
    - learning_rate: Scalar learning rate.
    - momentum: Scalar between 0 and 1 giving the momentum value.
    Setting momentum = 0 reduces to sgd.
    - velocity: A numpy array of the same shape as w and dw used to store a moving
    average of the gradients.
    """
    if config is None: 
        config = {}
        config.setdefault('learning_rate', learning_rate)
        config.setdefault('momentum', 0.9)
    v = config.get('velocity', np.zeros_like(w))

    next_w = None
    #############################################################################
    # Implement the momentum update formula. Store the updated value in         #
    # the next_w variable. You should also use and update the velocity v.       #
    #############################################################################
    v = config['momentum']*v - config['learning_rate']*dw
    next_w = w + v

    config['velocity'] = v

    return next_w, config



def rmsprop(x, dx, config=None):
    """
    Uses the RMSProp update rule, which uses a moving average of squared gradient
    values to set adaptive per-parameter learning rates.
    config format:
    - learning_rate: Scalar learning rate.
    - decay_rate: Scalar between 0 and 1 giving the decay rate for the squared
    gradient cache.
    - epsilon: Small scalar used for smoothing to avoid dividing by zero.
    - cache: Moving average of second moments of gradients.
    """
    if config is None: 
        config = {}
        config.setdefault('learning_rate', learning_rate)
        config.setdefault('decay_rate', 0.99)
        config.setdefault('epsilon', 1e-8)
        config.setdefault('cache', np.zeros_like(x))

    next_x = None
    #############################################################################
    # Implement the RMSprop update formula, storing the next value of x         #
    # in the next_x variable. Don't forget to update cache value stored in      #  
    # config['cache'].                                                          #
    #############################################################################
    config['cache'] = config['decay_rate'] * config['cache'] + (1 - config['decay_rate']) * dx**2
    next_x = x - config['learning_rate'] * dx / (np.sqrt(config['cache']) + config['epsilon'])

    return next_x, config


def adam(x, dx, config=None):
    """
    Uses the Adam update rule, which incorporates moving averages of both the
    gradient and its square and a bias correction term.
    config format:
    - learning_rate: Scalar learning rate.
    - beta1: Decay rate for moving average of first moment of gradient.
    - beta2: Decay rate for moving average of second moment of gradient.
    - epsilon: Small scalar used for smoothing to avoid dividing by zero.
    - m: Moving average of gradient.
    - v: Moving average of squared gradient.
    - t: Iteration number.
    """
    if config is None: 
        config = {}
        config.setdefault('learning_rate', learning_rate)
        config.setdefault('beta1', 0.9)
        config.setdefault('beta2', 0.999)
        config.setdefault('epsilon', 1e-8)
        config.setdefault('m', np.zeros_like(x))
        config.setdefault('v', np.zeros_like(x))
        config.setdefault('t', 0)

    next_x = None
    #############################################################################
    # Implement the Adam update formula, storing the next value of x in         #
    # the next_x variable. Don't forget to update the m, v, and t variables     #
    # stored in config.                                                         #
    #############################################################################
    config['t'] = config['t'] + 1
    config['m'] = config['beta1']*config['m'] + (1-config['beta1'])*dx
    config['v'] = config['beta2']*config['v'] + (1-config['beta2'])*(dx**2)

    mt = config['m'] / (1-config['beta1']**config['t'])
    vt = config['v'] / (1-config['beta2']**config['t'])

    next_x = x - config['learning_rate'] * mt / (np.sqrt(vt) + config['epsilon'])

    return next_x, config
```
